```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

Q5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a synthetic dataset with correlated features
np.random.seed(0)  # For reproducibility

# Create a base feature
x = np.random.normal(size=100)

# Generate correlated features
y1 = x + np.random.normal(scale=0.5, size=100)  # Correlated with x
y2 = x - np.random.normal(scale=0.5, size=100)  # Correlated with x
y3 = np.random.normal(size=100)  # Uncorrelated feature

# Combine into a DataFrame
data = pd.DataFrame({
    'Feature 1': x,
    'Feature 2': y1,
    'Feature 3': y2,
    'Feature 4': y3
})

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True, cbar=True)

# Add title
plt.title('Correlation Matrix Heatmap')

# Display the plot
plt.show()
```

Q1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x

```
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Prepare the output in the notebook
output_notebook()

# Generate x and y values
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.sin(x)  # Sine of x

# Create a Bokeh figure
p = figure(title="Sine Wave", x_axis_label='X', y_axis_label='Y')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="sin(x)", line_width=2, color="blue")

# Show the plot
show(p)
```

Q2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.

```
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np
import pandas as pd

# Prepare the output in the notebook
output_notebook()

# Generate random data
np.random.seed(42)  # For reproducibility
n = 100
x = np.random.rand(n) * 100  # Random x values from 0 to 100
y = np.random.rand(n) * 100  # Random y values from 0 to 100
sizes = np.random.randint(5, 50, n)  # Random sizes for the markers
colors = np.random.rand(n, 3)  # Random colors (RGB)

# Create a DataFrame for better handling (optional)
data = pd.DataFrame({
    'x': x,
    'y': y,
    'sizes': sizes,
    'colors': [f"rgb({int(c[0]*255)}, {int(c[1]*255)}, {int(c[2]*255)})" for c in colors]
})

# Create a Bokeh figure
p = figure(title="Random Scatter Plot", x_axis_label='X', y_axis_label='Y')

# Add a scatter renderer
p.scatter(x=data['x'], y=data['y'], size=data['sizes'], color=data['colors'], alpha=0.6)

# Show the plot
show(p)
```

Q3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
 fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
 counts = [20, 25, 30, 35]

```
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource

# Define the data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource for the data
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Configure Bokeh to display in the notebook
output_notebook()

# Create a new figure for the bar chart
p = figure(x_range=fruits, title="Fruit Counts", toolbar_location=None, tools="",
           y_axis_label="Count", x_axis_label="Fruit Type")

# Create the bar chart
p.vbar(x='fruits', top='counts', width=0.9, source=source, legend_field="fruits",
       line_color='white', fill_color=factor_cmap('fruits', palette=['#c9d9d3', '#718dbf', '#e84d60', '#ddb7b1'], factors=fruits))

# Customize the plot
p.y_range.start = 0
p.xgrid.grid_line_color = None
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Show the plot
show(p)
```

 Q4. Create a Bokeh histogram to visualize the distribution of the given data.
 data_hist = np.random.randn(1000)
 hist, edges = np.histogram(data_hist, bins=30)

```
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Prepare data
np.random.seed(25)
data_hist = np.random.randn(1000)  # Generate 1000 random numbers from a standard normal distribution

# Create histogram
hist, edges = np.histogram(data_hist, bins=30)

# Configure Bokeh to display in the notebook
output_notebook()

# Create a Bokeh figure
p = figure(title="Bokeh Histogram Example", background_fill_color="#fafafa")

# Add a quad glyph to create the histogram bars
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.7)

# Add labels
p.xaxis.axis_label = 'Value'
p.yaxis.axis_label = 'Frequency'

# Show the plot
show(p)
```

Q5. Create a Bokeh heatmap using the provided dataset.
 data_heatmap = np.random.rand(10, 10)
 x = np.linspace(0, 1, 10)
 y = np.linspace(0, 1, 10)
 xx, yy = np.meshgrid(x, y)

```
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColorBar, LinearColorMapper
import numpy as np
import pandas as pd

# Prepare the dataset
np.random.seed(42)  # For reproducibility
data_heatmap = np.random.rand(10, 10)  # 10x10 grid of random values
x = np.linspace(0, 1, 10)  # x values
y = np.linspace(0, 1, 10)  # y values
xx, yy = np.meshgrid(x, y)

# Convert the data into a DataFrame for easier plotting
data = pd.DataFrame({
    'X': xx.flatten(),
    'Y': yy.flatten(),
    'Value': data_heatmap.flatten()
})

# Configure Bokeh to display in the notebook
output_notebook()

# Create a color mapper
color_mapper = LinearColorMapper(palette="Viridis256", low=data['Value'].min(), high=data['Value'].max())

# Create a figure for the heatmap
p = figure(title="Bokeh Heatmap Example", x_axis_label='X', y_axis_label='Y',
           x_range=(x.min(), x.max()), y_range=(y.min(), y.max()), toolbar_location=None, tools="")

# Create the heatmap using rectangular glyphs
p.rect(x='X', y='Y', width=0.1, height=0.1, source=data,
       fill_color={'field': 'Value', 'transform': color_mapper}, line_color=None)

# Add a color bar to the right of the heatmap
color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
